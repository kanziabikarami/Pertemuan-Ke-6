```python
opsd_daily = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/opsd_germany_daily.csv')
opsd_daily.shape
```

```python
opsd_daily.info()
```

```python
opsd_daily['datetime']= pd.to_datetime(opsd_daily['Date'])
opsd_daily.info
```

```python
opsd_daily.drop(['Date'], axis=1, inplace=True)
opsd_daily.info()
```

```python
opsd_daily.head(3)
```

```python
opsd_daily.set_index('datetime', inplace=True)
opsd_daily.head(3)
```

```python
opsd_daily['Year'] = opsd_daily.index.year
opsd_daily['Month'] = opsd_daily.index.month
opsd_daily['Weekday'] = opsd_daily.index.weekday
opsd_daily.head()
```
